```python
print(hotelBookings.info(), '\n\n\n')
print('Info on null values in dataset:\n')
print(hotelBookings.isnull().sum())
```

```python
# Cleaning the dataset
substitutes = {'children':0.0, 'country':'Unknown', 'agent':0.0, 'company':0.0}
hotelBookings.fillna(substitutes, inplace = True)
```

```python
#meal has Undefined values set, that actually should be the same as SC, therefore they are set to SC
hotelBookings['meal'].replace(to_replace = 'Undefined', value = 'SC', inplace = True)
```

```python
noGuests = list(hotelBookings.loc[hotelBookings['adults']+hotelBookings['children'] + hotelBookings['babies'] == 0].index)
noGuests
```

```python
#hotel bookings that have no guests don't seem right hence will be taken out
hotelBookings.drop(hotelBookings.index[noGuests], inplace = True)
```

```python
#seperating resort and city hotels
resort = hotelBookings.loc[(hotelBookings['hotel'] == 'Resort Hotel')]
city = hotelBookings.loc[(hotelBookings['hotel'] == 'City Hotel')]
```
