```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from IPython.display import display
#display all columns, jupyter notebook by default will display first 10 and last 10 columns only
pd.set_option('display.max_columns', None)
```

```python
hotelBookings = pd.read_csv("D:\\databases\\hotel\\hotel_bookings.csv")
```

```python
hotelBookings
```

```python
print(hotelBookings.info(), '\n\n\n')
print('Info on null values in dataset:\n')
print(hotelBookings.isnull().sum())
```

```python
# Cleaning the dataset
substitutes = {'children':0.0, 'country':'Unknown', 'agent':0.0, 'company':0.0}
hotelBookings.fillna(substitutes, inplace = True)
```

```python
#meal has Undefined values set, that actually should be the same as SC, therefore they are set to SC
hotelBookings['meal'].replace(to_replace = 'Undefined', value = 'SC', inplace = True)
```

```python
noGuests = list(hotelBookings.loc[hotelBookings['adults']+hotelBookings['children'] + hotelBookings['babies'] == 0].index)
noGuests
```

```python
#hotel bookings that have no guests don't seem right hence will be taken out
hotelBookings.drop(hotelBookings.index[noGuests], inplace = True)
```

```python
#seperating resort and city hotels
resort = hotelBookings.loc[(hotelBookings['hotel'] == 'Resort Hotel')]
city = hotelBookings.loc[(hotelBookings['hotel'] == 'City Hotel')]
```

## 2. With the use of classification can I determine if a booking will be cancelled?

### How many of all the bookings were cancelled?

```python
cancelledReservations = pd.DataFrame(hotelBookings["is_canceled"].value_counts())
print(cancelledReservations)
```

```python
subs = {0 : 'notCancelled', 1: 'Cancelled'}
cancelledReservations.rename(index=subs,inplace = "True")
```

```python
print(cancelledReservations)
```

```python
#taking the values for cancelled and not cancelled
notCancelled = cancelledReservations.iat[0,0]
cancelled = cancelledReservations.iat[1,0]
labels = 'Cancelled', 'Not Cancelled'
sizes = [cancelled, notCancelled]
explode = [0.1,0]
#drawing the pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes,explode=explode, labels=labels, shadow=True, startangle = 90)
ax1.axis('equal')
plt.show
```

### how many different hotels?

```python
resortAmount = len(resort)
cityAmount = len(city)
```

```python
labels = 'Resort Hotel', 'City Hotel'
sizes = [resortAmount, cityAmount]
#pie chart for different hotels
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, shadow=True, startangle = 90)
ax1.axis('equal')
plt.show
```

it seems as though almost resort hotel and city hotel have a similar split to that of cancellations, lets see how cancellations split between hotel type

### cancellations per hotel type

```python
resortCancellations = len(resort[resort['is_canceled'] == 1])
resortNotCancellations =  len(resort[resort['is_canceled'] == 0])
cityCancellations =  len(city[city['is_canceled'] == 1])
cityNotCancellations = len(city[city['is_canceled'] == 0])
```

```python
cityLabels = 'Cancelled', 'Not Cancelled'
resortLabels = 'Cancelled', 'Not Cancelled'
citySizes = [cityCancellations,cityNotCancellations]
resortSizes = [resortCancellations, resortNotCancellations]

#the 2 different piecharts
fig1, ax1 = plt.subplots()
ax1.pie(citySizes, labels=cityLabels, shadow=True, startangle = 90)
ax1.set(aspect ="equal", title = 'City Cancellations')

fig2, ax2 = plt.subplots()
ax2.pie(resortSizes, labels=resortLabels, shadow=True, startangle = 90)
ax2.set(aspect ="equal", title = 'Resort Cancellations')
plt.show
```
